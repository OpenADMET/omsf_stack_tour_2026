# OpenADMET 2026 Demo

Hi all and welcome to the 2026 OpenADMET Demo! Here we will show an example of using our stack to do some training and finetuning on the [ExpansionRx dataset](https://huggingface.co/spaces/openadmet/OpenADMET-ExpansionRx-Challenge) that we recently ran a blind challenge on. 

Our aim is to show in practice how one might use the OpenADMET stack to approach a **tangible set of program data**, some of the relevant pitfalls and how to finetune to optimise performance on your own programs. 




## The ExpansionRx set 

Here we will use the ExpansionRx dataset as a test bed, the largest public ADMET dataset released to date from a real drug discovery program, representing a cumulative  > 7500 compounds across 9 ADME assays collected over ~5 years for Expansion's programs in RNA mediated and neurodegenerative diseases. This gives us a **realistic** program dataset to benchmark on. 


<img src="img/expansion.png" width="600">


## Setting the stage

Consider this scenario: you have just started a new drug program and you want to use a model to help you predict certain ADMET endpoints so you know which compounds to prioritize in assays. **Lets run through some approaches you might take**

* (1) Use **zero-shot models** - use models already available in the public domain
* (2) Training your own model on **public data** - leverage data from public sources yourself
* (3) Training a model on **public data and your own data**. - combine public data with your own data

We will run through examples of each, see how well they work using the ExpansionRx canonical test-train split. 

## Zero-Shot models

Something you might try first is to use some **off-the shelf** pre-existing models. These models are in the public domain (often with webservers) and are the lowest overhead way to get an ADMET prediction. We will call this **zero-shot** prediction as these public models are trained on effectively unrelated data curated by the authors of these models. 

How likely is this to work? 

We will cover two popular webserver implementations 

* ADMET-AI, a ChemProp v1 GNN enhanced with RDKit descriptors trained on data from the Therapeutic Data Commons (TDC), and comes from a [web-based tool](https://admet.ai.greenstonebio.com/?ref=openadmet.ghost.io).

* ADMETLab 3.0 similarly is a ChemProp v1 model trained on 400,000 molecules curated from ChEMBL, PubChem and OCHEM. Its available at this web server [here](https://admetlab3.scbdd.com/?ref=openadmet.ghost.io).

## Lets start with an easy endpoint: LogD

## Can we do a bit better ourselves? - training a model using Anvil

Much focus in our field has focused on models and architectures relative to gathering required data to make models better. However, it's worth testing whether different approaches to training these types of models will produce different results. 

With this in mind OpenADMET has trained a series of public-data models that we have on our HuggingFace. These are trained mostly on data curated from ChEMBL and serve as our **best-effort zero-shot models** trained with our stack. We will run through an example of how to do this below. 



```python
# read pre-curated data from our `data-catalogs` repo, more detail about reference code used to curate in our demos repo
# https://demos.openadmet.org

import intake 
cat = intake.open_catalog(
    "https://github.com/OpenADMET/data-catalogs/raw/refs/heads/main/catalogs/activities/ChEMBL_LogD/CATALOG_ChEMBL35_LogD.yaml"
)
```

```python
list(cat.entries)
```

```python
chembl_logd = cat["LogD_aggregated"].read()
```

```python
chembl_logd.reset_index(drop=True).head()
```

```python
!openadmet anvil --recipe-path anvil_logd.yaml --output-dir logd
```

```python
!openadmet predict \
    --input-path data/exprx_test.csv \
    --input-col OPENADMET_CANONICAL_SMILES \
    --model-dir logd \
    --output-csv chembl_predictions.csv \
    --accelerator cpu
```

## Zero shot models don't work well

Zero shot models do not work well at all. While we see some slight improvement in our implementation relative to other public-data implmentations, the conclusion remains **data is the key to good performance**
